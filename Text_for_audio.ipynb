```python
from gtts import gTTS as gtts

texto = "Do you speak english ?"

language = "en"

gtts_object = gtts(text = texto,
               lang= language,
               slow = False)
gtts_object.save("/content/gtts.wav")
```

```python
from googletrans import Translator

texto_portugues = "Você sabe falar em português?"

translator = Translator()
traducao = translator.translate(texto_portugues, dest = "en")


print(traducao.text)
```

```python

tts_object = gtts(text =  traducao.text,
                  lang = "en",
                  slow = False)
tts_object.save("/content/traducao.wav")
```

```python
from IPython.display import Audio

Audio("/content/traducao.wav")
```
